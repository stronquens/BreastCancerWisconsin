<h1>Breast Cancer of Wisconsin</h1>

<center><h2>Importing dataframes</h2></center>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

A) Breast Cancer Wisconsin Dataframe

```python
names = ['Code number', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 
         'Marginal Adhesion', 'Single Epithelial Cell Size', 'Bare Nuclei', 'Bland Chromatin', 'Normal Nucleoli',
        'Mitoses', 'Class']
breastCancer = pd.read_csv("./data/breast-cancer-wisconsin.data", sep=",", names=names)
breastCancer.head()
# Class: 2=benign or 4=malignant.
# rest 1-10
```

B) Wisconsin Diagnostic Breast Cancer (WDBC) dataframe

```python
wdbcNames = ['id', 'diagnosis']
realValedFeatures = ['radius', 'texture', 'perimeter', 'area', 'smoothness', 'compactness', 'concavity', 
                     'concave_points', 'symmetry', 'fractal_dimension']
# Tres mediciones para cada característica, las características se dividen en 
# mean, standard error, and Worst Radius
featureMean = [s + '_mean' for s in realValedFeatures]
featureSd = [s + '_sd' for s in realValedFeatures]
featureWorst = [s + '_worst' for s in realValedFeatures]

wdbcNames = wdbcNames + featureMean + featureSd + featureWorst
```

```python
wdbc = pd.read_csv("./data/wdbc.data", sep=",", names=wdbcNames)
wdbc['diagnosis'] = wdbc['diagnosis'].astype('category')
wdbc.head()
# diagnosis: B = benign, M = malignant
```

```python
print("Shape: ", wdbc.shape)
print("Missing data: ")
#print(np.sum(wdbc == 'none'))
print(wdbc.info())
```

C) Wisconsin Prognostic Breast Cancer (WPBC)

```python
wpbcNames = ['id', 'outcome', 'time']
realValedFeatures = ['radius', 'texture', 'perimeter', 'area', 'smoothness', 'compactness', 'concavity', 
                     'concave_points', 'symmetry', 'fractal_dimension']
# Tres mediciones para cada característica, las características se dividen en 
# mean, standard error, and Worst Radius
featureMean = [s + '_mean' for s in realValedFeatures]
featureSd = [s + '_sd' for s in realValedFeatures]
featureWorst = [s + '_worst' for s in realValedFeatures]

wpbcNames = wpbcNames + featureMean + featureSd + featureWorst
```

```python
wpbc = pd.read_csv("./data/wdbc.data", sep=",", names=wpbcNames)
wpbc.head()
# outcome: R = recurrent, N = nonrecurrent
# Time (recurrence time if field 2 = R, disease-free time iffield 2 = N)
```

<center><h2> Exploratory Analisis</h2></center>

<h3>a) Duplicates and missing data</h3>

```python
print("Duplicates:")
print("Shape: ", wdbc.shape)
print("Shape without duplicates:", wdbc.drop_duplicates().shape)

print("Missing data: ")
print(np.sum(wdbc == 'none'))
```

<h3>b) Correlations</h3>

```python
from sklearn import preprocessing
ldiag = preprocessing.LabelEncoder()
wdbc['diagnosis'] = ldiag.fit_transform(wdbc['diagnosis'])
```

```python
correlation_mat = wdbc.iloc[:,1:12].corr()

fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(correlation_mat, annot=True, linewidths=.5, ax=ax, cmap="Blues")
plt.title("Correlation matrix of Breast Cancer data")
plt.xlabel("cell nucleus features")
plt.ylabel("cell nucleus features")
plt.show()
```

```python
correlation_mat = wdbc.loc[:,['diagnosis']+featureSd].corr()

fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(correlation_mat, annot=True, linewidths=.5, ax=ax, cmap="Blues")
plt.title("Correlation matrix of Breast Cancer data")
plt.xlabel("cell nucleus features")
plt.ylabel("cell nucleus features")
plt.show()
```

```python
correlation_mat = wdbc.loc[:,['diagnosis']+featureWorst].corr()

fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(correlation_mat, annot=True, linewidths=.5, ax=ax, cmap="Blues")
plt.title("Correlation matrix of Breast Cancer data")
plt.xlabel("cell nucleus features")
plt.ylabel("cell nucleus features")
plt.show()
```

```python
correlation_mat = wdbc.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values(kind="quicksort")
print(sorted_pairs)
```

```python
negative_pairs = sorted_pairs[sorted_pairs < 0]

print(negative_pairs)
```

```python
strong_pairs = sorted_pairs[abs(sorted_pairs) > 0.5]

print(strong_pairs)
```
